```python
# final function of word_count
def word_count(source):
    """source: the input file name in current folder
    return a list of sorted tuples by count"""
    from pyspark import SparkConf, SparkContext
    from operator import add
    conf = SparkConf().setMaster('local').setAppName('testApp')
    sc = SparkContext(conf = conf)
    lines = sc.textFile(source)
    counts = lines.flatMap(lambda line: line.split(' ')).map(lambda word: (word, 1)).reduceByKey(add)
    res = counts.collect()
    sorted_res = sorted(res, key=lambda x: x[1], reverse=True)
    sc.stop()
    return sorted_res
```

```python
temp=word_count('test.txt')
```

```python
print(type(sorted_res))
```

```python
for (word, count) in temp:
    print(word, count)
```
